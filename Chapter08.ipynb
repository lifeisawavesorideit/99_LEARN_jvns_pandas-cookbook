# 8.1 Parsing Unix timestamps

```python
csv_path = 'https://raw.githubusercontent.com/lifeisawavesorideit/99_LEARN_jvns_pandas-cookbook/master/data/popularity-contest'
```

```python
df = pd.read_csv(csv_path, sep=' ')[:-1]
df.columns = ['atime', 'ctime', 'package-name', 'mru-program', 'tag']
df 
```

```python
df.dtypes
```

```python
df['atime'] = df['atime'].astype(int)
df['ctime'] = df['ctime'].astype(int)
```

```python
df.dtypes
```

```python
df['atime'] = pd.to_datetime(df['atime'], unit='s')
df['ctime'] = pd.to_datetime(df['ctime'], unit='s')
```

```python
df
```

```python
df = df[df['atime'].dt.year > 1970]
```

```python
df
```

```python
nonlibraries = df[~df['package-name'].str.contains('lib')]
```

```python
nonlibraries.sort_values('ctime', ascending=False)
```
