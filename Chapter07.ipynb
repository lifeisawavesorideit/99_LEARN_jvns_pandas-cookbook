# 7.2 Fixing the nan values and string/float confusion


```python
na_values = ['NO CLUE', 'N/A', '0']
df = pd.read_csv(csv_path, sep=',', na_values=na_values, dtype={'Incident Zip': str})
```

# 7.3 What's up with the dashes?


```python
rows_with_dashes = df['Incident Zip'].str.contains('-').fillna(False)
```

```python
rows_with_dashes
```

```python
df[rows_with_dashes]
```

```python
long_zip_codes = df['Incident Zip'].str.len()>5
```

```python
long_zip_codes
```

```python
df['Incident Zip'][long_zip_codes].unique()
```

```python
df['Incident Zip'] = df['Incident Zip'].str.slice(0, 5)
```

```python
df[df['Incident Zip'] == '00000']
```

```python
zero_zips = df['Incident Zip'] == '00000'
```

```python
df.loc[zero_zips, 'Incident Zip'] = np.nan
```

```python
unique_zips = df['Incident Zip'].unique()
```

```python
unique_zips
```

```python
zips = df['Incident Zip']
is_close = zips.str.startswith('0') | zips.str.startswith('1')
is_far = ~(is_close) & zips.notnull()
```

```python
is_close
```

```python
is_far
```
